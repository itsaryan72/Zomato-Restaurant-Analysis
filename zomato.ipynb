```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")

#1. Bar Chart for Top 10 Cities by Restaurant Count
plt.figure(figsize=(12,6))
top_cities = df['City'].value_counts().head(10)
sns.barplot(x=top_cities, y=top_cities.index, color='skyblue')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()
```

```python
#2.Histogram: Aggregate Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='green')
plt.title('Distribute of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#3. Box Plot: Price Range vs Aggregate Rating
plt.figure(figsize=(10,6))
sns.boxplot(x="Price range", y="Aggregate rating", data=df, color="orange")
plt.title('Price Range cs Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#4. Bar Chart: Top 10 Most Popular Cuisines
plt.figure(figsize=(12,6))
top_cuisines = df['Cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, color="lightcoral")
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

```python
#5. Scatter Plot: Restaurant Locations
plt.figure(figsize=(12,6))
sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='City', legend=False, alpha=0.9)
plt.title('Geographic Distribution of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#6. Heatmap: Correlation Matrix
plt.figure(figsize=(12,6))
corr = df[['Aggregate rating', 'Votes', 'Average Cost for two', 'Price range']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
